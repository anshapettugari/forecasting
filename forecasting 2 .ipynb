```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  
```

```python
cola=pd.read_excel("CocaCola_Sales_Rawdata (1).xlsx")
cola
```

```python
pd.get_dummies(cola,prefix=None,prefix_sep='_',dummy_na=False,columns=None,sparse=False,drop_first=False,dtype=None)
```

```python
from pmdarima import auto_arima
import warnings
warnings.filterwarnings("ignore")
```

```python
stepwise_fit=auto_arima(cola["Sales"],trace=True,suppress_warnings=True)
stepwise_fit.summary()
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
print(cola.shape)
train=cola.iloc[:-10]
test=cola.iloc[-10:]
print(train.shape,test.shape)
```

```python
model=ARIMA(train["Sales"],order=(4,1,0))
model=model.fit()
model.summary()
```

```python
start=len(train)
end=len(train)+len(test)-1
pred=model.predict(start=start,end=end,typ="levels")
print(pred)
```

```python
pred.plot(legend=True)
test["Sales"].plot(legend=True)
```

```python
test.mean()
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rmse=sqrt(mean_squared_error(pred,test["Sales"]))
print(rmse)
```
